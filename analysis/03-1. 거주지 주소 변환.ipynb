```python
# 혹시 모를 일을 대비해 백업
apt_raw = apt.copy()
multi_raw = multi.copy()

# 컬럼 이름 공백 제거 (중요)
apt.columns = apt.columns.str.strip()
multi.columns = multi.columns.str.strip()

# 필요한 컬럼만 추출 + 컬럼명 통일
apt = apt[['시군구', '단지명', '도로명']].copy()
multi = multi[['시군구', '건물명', '도로명']].copy()
multi.rename(columns={'건물명': '단지명'}, inplace=True)  # 아파트와 맞추기

# 주소 컬럼 생성 (도로명만 사용)
apt['주소'] = apt['시군구'].astype(str).str.strip() + ' ' + apt['도로명'].astype(str).str.strip()
multi['주소'] = apt['시군구'].astype(str).str.strip() + ' ' + apt['도로명'].astype(str).str.strip()

# 확인
print("정리된 아파트 데이터:")
display(apt.head())

print("정리된 연립/다세대 데이터:")
display(multi.head())
```

```python
import pandas as pd
import requests
import time
from tqdm import tqdm
import urllib.parse

#  카카오 API 설정
KAKAO_API_KEY = input("Kakao REST API 키를 입력하세요: ").strip()

def kakao_geocode(address):
    try:
        query = urllib.parse.quote(address)
        url = f'https://dapi.kakao.com/v2/local/search/address.json?query={query}'
        headers = {'Authorization': f'KakaoAK {KAKAO_API_KEY}'}
        res = requests.get(url, headers=headers)
        if res.status_code != 200:
            print(f"[ERROR {res.status_code}] {address}")
            return None, None
        result = res.json()['documents']
        if result:
            return result[0]['y'], result[0]['x']
        else:
            print(f"[NO RESULT] {address}")
            return None, None
    except Exception as e:
        print(f"[EXCEPTION] {address}: {e}")
        return None, None


# 캐시로 중복요청 방지
geocoded_cache = {}

def safe_geocode(address):
    if address in geocoded_cache:
        return geocoded_cache[address]
    lat, lon = kakao_geocode(address)
    geocoded_cache[address] = (lat, lon)
    time.sleep(0.2)  # 카카오 속도 제한 회피
    return lat, lon
```

```python
# 아파트 데이터 주소 좌표 변환
apt_coords = []
for addr in tqdm(apt['주소'], desc='아파트 주소 변환 중'):
    lat, lon = safe_geocode(addr)
    apt_coords.append((lat, lon))

apt['위도'] = [c[0] for c in apt_coords]
apt['경도'] = [c[1] for c in apt_coords]
```

```python
print(len(apt_coords))  # 19565면 완료
```

```python
# 연립다세대 데이터 주소 좌표 변환
multi_coords = []
for addr in tqdm(multi['주소'], desc='연립다세대 주소 변환 중'):
    lat, lon = safe_geocode(addr)
    multi_coords.append((lat, lon))

multi['위도'] = [c[0] for c in multi_coords]
multi['경도'] = [c[1] for c in multi_coords]
```
